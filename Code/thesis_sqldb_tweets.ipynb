```python
import os
import re
import json
import sqlite3

# Define the list of ticker symbols
#tickers = ['MSFT', 'GOOG', 'AXISBANK.NSE','BAJFINANCE.NSE','BANKBARODA.NSE', 'FEDRALBNK.NSE', 'HDFCBANK.NSE', 'ICICIBANK.NSE', 'IDFCFIRSTB.NSE', 'INDUSINDBK.NSE', 'INFY.NSE', 'KOTAKBANK.NSE', 'PNB.NSE', 'RBLBANK.NSE', 'SBIN.NSE', 'TCS.NSE', 'WIPRO.NSE']
tickers = ['MSFT', 'GOOG', 'AAPL', 'NVDA']
tickers = ['TSLA']
# Create SQLite3 database if it doesn't exist
db_file = 'stock_data.db'
conn = sqlite3.connect(db_file)
c = conn.cursor()

# Loop through each ticker
for ticker in tickers:
    print(f"Processing ticker {ticker}")
    ticker_table_name = f'tweets_{ticker.replace(".", "_")}'  # Create table name based on ticker symbol
    
    # Drop the table if it already exists
    c.execute(f'DROP TABLE IF EXISTS {ticker_table_name}')
    
    # Create the table with an identity column
    c.execute(f'''CREATE TABLE {ticker_table_name}
                 (ID INTEGER PRIMARY KEY AUTOINCREMENT, Symbol TEXT, Body TEXT, CreatedBy TEXT, CreatedAt TEXT, MessageID INTEGER, Sentiment TEXT)''')

    ticker_folder = os.path.join('..\stocktwits', ticker)
    if os.path.exists(ticker_folder):
        # Get the list of JSON files in the ticker folder
        filenames = [filename for filename in os.listdir(ticker_folder) if filename.endswith('.json')]
        
        # Sort filenames based on the number extracted from the filename
        if (ticker.find(".NSE")>0):
            filenames.sort(key=lambda x: int(re.search(r'output-\w+\.\w+-(\d+)\.json', x).group(1)))
        else:
            filenames.sort(key=lambda x: int(re.search(r'output-\w+-(\d+)\.json', x).group(1)))
        
        # Iterate through sorted filenames
        for filename in filenames:
            print(f"Processing file {filename}")
            filepath = os.path.join(ticker_folder, filename)
            # Check if the file is empty
            if os.path.getsize(filepath) == 0:
                continue
            with open(filepath, 'r') as file:
                # Load JSON data from file
                tweets = json.load(file)
                for tweet in tweets:
                    # Extract tweet fields
                    symbol = ticker #tweet['Symbol']
                    body = tweet['Body']
                    created_by = tweet['CreatedBy']
                    created_at = tweet['CreatedAt']
                    message_id = tweet['MessageID']
                    # Extract sentiment
                    sentiment = None
                    if tweet['Sentiment'] is not None:
                        sentiment = tweet['Sentiment'].get('basic')
                    
                    # Insert data into the table
                    c.execute(f"INSERT INTO {ticker_table_name} (Symbol, Body, CreatedBy, CreatedAt, MessageID, Sentiment) VALUES (?, ?, ?, ?, ?, ?)",
                              (symbol, body, created_by, created_at, message_id, sentiment))

# Commit changes and close connection
conn.commit()
conn.close()
```
